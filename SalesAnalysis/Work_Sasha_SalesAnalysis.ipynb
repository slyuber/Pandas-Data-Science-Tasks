### Task 1: Merger all data files into one

```python
## Use ".\" to signifiy current dirrectory

df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
df.head(10)
```

#### Make a temp df while we play around

```python
df_temp = df
df_temp.head(10)
```

#### Get a month column going

```python
df_temp["Month"] = df_temp["Order Date"].str[:2]
df_temp.head(10)
```

```python
## Lets check to see if we have any abd data
df_temp.loc[df_temp["Month"] != "04"]
```

#### Drop all rows containing all NaN values 

```python
df_temp = df_temp.dropna(how='all')
df_temp.loc[df_temp["Month"] != "04"]
```

#### Get rid of Order or "Or" in month column by removing bad data

```python
df_temp.loc[df_temp["Order ID"].str.contains('Order')]
```

```python
## Get rid of anything containing "Or"
df_temp = df_temp.loc[~df_temp["Order ID"].str.contains('Order')]
df_temp.head(10)
```

```python
## Check to see if data is clean

df_temp.loc[~df_temp["Month"].str.contains('04')]
```
